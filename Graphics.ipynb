```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
distances = np.linspace(0.2, 5, 1000)
f_spring = 2 * np.log(distances / 1)
f_rep = -1 / distances ** 2
f_attr = f_spring - f_rep
ax.plot(distances, f_spring, label='spring')
ax.plot(distances, f_rep, label='repulsive')
ax.plot(distances, f_attr, label='attractive')
ax.scatter(1, 0, label='$\it{L}$')
ax.set_xlabel('Расстояние')
ax.set_ylabel('Сила')
plt.legend();
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
distances = np.linspace(0.2, 5, 1000)
f_rep = -(1.5 ** 2) / distances
f_attr = distances ** 2 / 1.5
f_central = 2 * (f_rep + f_attr)
ax.plot(distances, f_rep, label='repulsive')
ax.plot(distances, f_attr, label='attractive')
ax.plot(distances, f_central, label='central')
ax.scatter(1.5, 0, label='$\it{L=1.5}$')
ax.set_xlabel('Расстояние')
ax.set_ylabel('Сила')
plt.legend();
```
